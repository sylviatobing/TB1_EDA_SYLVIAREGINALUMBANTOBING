```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import norm
import statsmodels.formula.api as smf
```

```python
df = pd.read_csv('/content/drive/MyDrive/Visualisasi_data/dataset/employee.csv')
df.head()
```

```python
df.info() #table info
```

```python
#1. Drop coloumn yang tidak diperlukan
df = df.drop(['EmployeeNumber','Department','Over18','StandardHours','DailyRate','DailyRate','Education','EducationField','HourlyRate','JobLevel','MonthlyRate','NumCompaniesWorked','Over18','StandardHours','StockOptionLevel','TrainingTimesLastYear'], axis=1)
```

```python
# Creating numerical variables 
num_cols = ['Age', 'DistanceFromHome', 'MonthlyIncome', 'PercentSalaryHike', 'TotalWorkingYears',
          'YearsAtCompany', 'YearsInCurrentRole', 'YearsSinceLastPromotion', 'YearsWithCurrManager']

# Creating categorical variables 
cat_cols = ['Attrition', 'OverTime', 'BusinessTravel', 'JobSatisfaction', 
            'EnvironmentSatisfaction', 'WorkLifeBalance', 'Gender', 'PerformanceRating', 'JobInvolvement',
            'JobRole', 'MaritalStatus', 'RelationshipSatisfaction']
```

```python
#Membuat histograms
df[num_cols].hist(figsize = (15, 15))
plt.tight_layout()
plt.show()
```

```python
#Membuat Boxplot

for column in num_cols: sns.catplot(x=
'Attrition'
, y=column, kind=
'box'
, data=df, hue=
'Attrition'
)
```

```python
# Generate the boxplots in a tiled layout
sns.set_style('whitegrid')  

# Set the style
g = sns.catplot(x='Attrition', y='value', kind='box', data=df.melt(id_vars='Attrition', value_vars=num_cols), hue='Attrition', col='variable', height=4, aspect=.7, sharey=False)
g.set_xticklabels(rotation=0)  
# Rotate x-axis labels for better visibility
```

```python
#Melihat statistik kolom
df[num_cols].describe().T
```

```python
#Jawaban
#identifikasi nilai upper whisker dan lower whisker dari boxplot tiap kolom serta menghitung count & proportion nya
#multivariate attrition
for column in num_cols:
    # Generate the boxplot
    bp = sns.boxplot(x='Attrition', y=column, data=df)

    # Calculate the upper and lower whisker values
    upper_whisker = bp.lines[3].get_ydata()[1]
    lower_whisker = bp.lines[2].get_ydata()[1]

    # Calculate the IQR
    q1, q3 = np.percentile(df[column], [25, 75])
    iqr = q3 - q1

    # Calculate the outlier thresholds
    upper_threshold = q3 + 1.5*iqr
    lower_threshold = q1 - 1.5*iqr

    # Calculate the count and proportion of outliers
    outliers_count = len(df[(df[column] > upper_threshold) | (df[column] < lower_threshold)])
    outliers_proportion = outliers_count / len(df)

    # Print the results
    print(f'Boxplot for {column}:')
    print(f'Upper whisker value = {upper_whisker:.2f}, Lower whisker value = {lower_whisker:.2f}')
    print(f'Number of outliers = {outliers_count}, Proportion of outliers = {outliers_proportion:.3f}')
    print('')
    
    # Show the plot
    plt.show()
```

Identifikasi hal yang menarik


1.   Distribusi usia mendekati distribusi normal, dengan sebagian besar karyawan berusia antara 25 dan 50 tahun.
2.   DistanceFromHome juga memiliki distribusi condong ke kanan, yang berarti sebagian besar karyawan tinggal dekat dengan tempat kerja tetapi ada beberapa yang tinggal lebih jauh.
3.   MonthlyIncome dan TotalWorkingYears condong ke kanan, menunjukkan bahwa sebagian besarkaryawan berada di posisi level masuk/menengah di organisasi.
4.   Persentase kenaikan gaji condong ke kanan, yang berarti karyawan kebanyakan mendapatkan kenaikan gaji 
5.   Distribusi variabel YearsAtCompany menunjukkan proporsi yang baik dari pekerja dengan pengalaman 10 tahun atau lebih, menunjukkan adanya jumlah karyawan setia kepada perusahaan
6.   Distribusi YearsInCurrentRole memiliki tiga puncak di 0, 2, dan 7. Ada beberapa karyawan yang bahkan tetap berada di posisi yang sama selama 15 tahun atau lebih
7.   Distribusi variabel YearsSinceLastPromotion menunjukkan bahwa beberapa karyawan belum menerima kenaikan jabatan dalam 10-15 tahun namun masih bekerja dlam perusahaan
